### Format Results

```python
def twdf(results):
    idlist = [tweet.id for tweet in results]
    df = pd.DataFrame(idlist, columns = ["id"])
    # tweet data
    df["text"] = list(map(lambda x: x.text, results))
    df["created_at"] = [tweet.created_at for tweet in results]
    df["retweet_count"] = [tweet.retweet_count for tweet in results]
    df["favorite_count"] = [tweet.favorite_count for tweet in results]
    df["source"] = [tweet.source for tweet in results]
    df["country"] = list(map(lambda tw: tw.place.country if tw.place != None else None, results))
    # User Data
    df["user_id"] = [tweet.author.id for tweet in results]
    df["user_screen_name"] = [tweet.author.screen_name for tweet in results]
    df["user_name"] = [tweet.author.name for tweet in results]
    df["user_created_at"] = [tweet.author.created_at for tweet in results]
    df["user_followers_count"] = [tweet.author.followers_count for tweet in results]
    df["user_friends_count"] = [tweet.author.friends_count for tweet in results]
    df["user_location"] = [tweet.author.location for tweet in results]
    return df

kanyedf = twdf(kanye)
kanyedf['text'] = kanyedf['text'].astype(str)

westdf = twdf(west)
westdf['text'] = westdf['text'].astype(str)
display(kanyedf.tail())
display(westdf.tail())

```

### Clean up tweets

```python
kanyedf['text'] = kanyedf['text'].str.replace('http\S+|www.\S+', '', case=False)
kanyedf['text'] = kanyedf['text'].str.replace('@\S+', '', case=False)
kanyedf['text'] = kanyedf['text'].str.replace('#', '')
kanyedf['text'] = kanyedf['text'].str.replace('RT ', '')
display(kanyedf.head())
```

```python
westdf['text'] = westdf['text'].str.replace('http\S+|www.\S+', '', case=False)
westdf['text'] = westdf['text'].str.replace('@\S+', '', case=False)
westdf['text'] = westdf['text'].str.replace('#', '')
westdf['text'] = westdf['text'].str.replace('RT ', '')
display(westdf.head())
```
